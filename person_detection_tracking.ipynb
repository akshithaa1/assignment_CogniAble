```python
import cv2
import torch
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

# Load YOLO model (you can also use a smaller or different model like 'yolov8n.pt')
model = YOLO('yolov8m.pt')

# Initialize DeepSORT object
tracker = DeepSort(max_age=50, nn_budget=100)  # You can tune the max_age and nn_budget values

def process_video(video_path, output_path):
    cap = cv2.VideoCapture(video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Define codec and output file
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Run YOLO inference
        results = model.predict(source=frame, save=False, imgsz=640, conf=0.3)

        # Prepare detections for DeepSORT (list of [[x1, y1, x2, y2], confidence])
        detections = []
        for result in results:
            boxes = result.boxes
            for bbox in boxes:
                x1, y1, x2, y2 = map(int, bbox.xyxy[0].cpu().numpy())  # Convert to int
                confidence = bbox.conf[0].item()  # Convert confidence to float
                class_id = int(bbox.cls[0])

                # Only track 'person' class (class_id == 0)
                if class_id == 0:  # 'person' class in COCO dataset
                    # Append detection as [[x1, y1, x2, y2], confidence]
                    detections.append([[x1, y1, x2, y2], confidence])

        # Update tracks using DeepSORT
        tracks = tracker.update_tracks(detections, frame=frame)

        # Draw bounding boxes and unique IDs for tracked objects
        for track in tracks:
            if not track.is_confirmed():
                continue
            track_id = track.track_id
            ltrb = track.to_ltrb()  # left-top-right-bottom bounding box coordinates
            x1, y1, x2, y2 = map(int, ltrb)

            # Draw bounding box and label with track ID
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f"ID {track_id}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

        # Write the processed frame with detections and tracking information to the output video
        out.write(frame)
        torch.cuda.empty_cache()  # Clear cache to prevent memory overload

    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Example usage
process_video("/content/ABA Therapy： Daniel - Communication [V9YDDpo9LWg].webm", "output_video_with_tracking.mp4")
```

```python
import cv2
import torch
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

# Load the YOLOv8 model
model = YOLO('yolov8l.pt')

# Initialize DeepSORT tracker with optimized parameters
tracker = DeepSort(max_age=45, nn_budget=150, nms_max_overlap=0.7)

def process_video(video_path, output_path):
    cap = cv2.VideoCapture(video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Define codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Run YOLOv8 inference
        results = model.predict(source=frame, save=False, imgsz=1280, conf=0.3, iou=0.5)

        # Prepare detections for DeepSORT tracker
        detections = []
        for result in results:
            boxes = result.boxes
            for bbox in boxes:
                x1, y1, x2, y2 = map(int, bbox.xyxy[0].cpu().numpy())  # Convert to integer coordinates
                confidence = bbox.conf[0].item()  # Convert confidence score to float
                class_id = int(bbox.cls[0])

                # Only track 'person' class (class_id == 0)
                if class_id == 0:  # 'person' class in COCO dataset
                    # Calculate bounding box size (to differentiate between therapist and child)
                    box_area = (x2 - x1) * (y2 - y1)
                    if box_area > (frame_width * frame_height * 0.1):  # Assume therapist is larger
                        label = "Therapist"
                    else:
                        label = "Child"
                    # Add the detection and label to the list
                    detections.append([[x1, y1, x2, y2], confidence, label])

        # Update tracks using DeepSORT tracker
        tracks = tracker.update_tracks([[d[0], d[1]] for d in detections], frame=frame)

        # Draw bounding boxes and labels on the frame
        for track in tracks:
            if not track.is_confirmed():
                continue

            track_id = track.track_id
            ltrb = track.to_ltrb()  # left-top-right-bottom bounding box coordinates
            x1, y1, x2, y2 = map(int, ltrb)

            # Find the corresponding label from the detections or set default label if not found
            label = next((d[2] for d in detections if d[0] == [x1, y1, x2, y2]), "Person")

            # Draw bounding box and track ID label
            color = (0, 255, 0) if label == "Therapist" else (255, 0, 0)  # Green for Therapist, Blue for Child
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, f"{label} (ID {track_id})", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

        # Write the processed frame to the output video
        out.write(frame)

        # Clear GPU memory
        torch.cuda.empty_cache()

    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Example usage
process_video("/content/output_video.mp4", "final_output_therapist_child.mp4")
```
